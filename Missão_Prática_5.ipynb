#4) Manipular os dados para visualizá-los em uma nuvem de palavras

````python
from wordcloud import WordCloud

texto ='um,um,um,dois,dois,um,tres,quatro,testando' 
nuvem_palavras = WordCloud(background_color='black',
                           width=800,height=400).generate(texto)
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis("off")
nuvem_palavras.to_file("Nuvem de palavras.png")
````

#5) Implementar um programa para gerar dados com nomes de pessoas e respectivas pontuações, gravar em um arquivo, recuperar os dados do arquivo, visualizar os dados das pontuações das pessoas em um histograma e em uma nuvem de palavras.

````python
try:
    from num2words import num2words
except:
    !pip install num2words
    from num2words import num2words
try:
    from wordcloud import WordCloud
except:
    !pip install wordcloud
    from wordcloud import WordCloud
try:
    from faker import Faker
except:
    !pip install faker
    from faker import Faker
import random, ast
import matplotlib.pyplot as plt

#Gerando o objeto fake usando a classe Faker 
fake = Faker('pt_BR')

#Gera um dicionário com uma quantidade determinada de nomes aleatórios e pontos entre 0 e 20
quantidade = 50
lista = {fake.name(): random.randint(0,20) for i in range(quantidade)}

#Salva o arquivo txt
open('nomes.txt', 'w').write(str(lista))

#Lê o arquivo txt
lista = ast.literal_eval(open('nomes.txt', 'r').read())

#Gerando o Histograma
lista = [lista[i] for i in lista]
maior = 0
for i in lista:
    vezes = lista.count(i)
    if vezes > maior:
        maior = vezes

num_bins = 20
plt.hist(lista, num_bins, facecolor='yellow', alpha=0.8, edgecolor='#000', linewidth=1.2)
plt.xlabel('Notas')
plt.ylabel('Quantidade')
plt.title('Notas dos alunos')
plt.gca().yaxis.grid(True, color='#000', linestyle='-', linewidth=1)
plt.xticks([i for i in range(21)])
plt.xlim([0, 20])
plt.yticks([i for i in range(maior + 2)]) #Valores
plt.ylim([0, maior + 1]) #Quantidade
plt.show()

#Gera a nuvem de palavras
lista = ','.join([num2words(i, lang='pt') for i in lista])
nuvem_palavras = WordCloud(background_color='#f0f0f0').generate(lista)
plt.imshow(nuvem_palavras, interpolation='bilinear')
pass
````

# Qual a importância de visualizar os dados em um histograma?


O histograma é importante para a análise estatística de diferentes eventos. Ele pode ser utilizado na medição de grandezas físicas, como: peso, altura, comprimento, temperatura, pressão, volume etc. Também pode ser útil no acompanhamento de ocorrências em geral, como: quantidade de telefonemas, e-mails, reclamações, leads, nível de satisfação com a empresa, e por aí vai.
Alguns dos benefícios são: Facilitar a visualização de um volumoso grupo de dados; Comparar grandezas; Facilitar a comunicação das informações.

# O que significa a visualização por nuvem de palavras?

Uma nuvem de palavras é um recurso gráfico (usado
principalmente na internet) para descrever os termos
mais frequentes de um determinado texto. O tamanho da
fonte em que a palavra é apresentada é uma função da
frequência da palavra no texto: palavras mais frequentes
são desenhadas em fontes de tamanho maior, palavras
menos frequentes são desenhadas em fontes de tamanho menor.
